```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
```

```python
print(type(model.names), len(model.names))

print(model.names)
```

```python
model.train(data = '/content/HAN/HAN.yaml', epochs=100, patience=100, batch=32, imgsz=416)
```
